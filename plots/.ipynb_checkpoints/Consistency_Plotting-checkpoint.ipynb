```python
df = pd.read_csv("../data/consistency/outdegree_centrality/consistency_SPLIT_individual_raw_Head-Head_Male_FCvsCC_outdegree_centrality.csv")
# print(df.head)
df_CC_1 = df.loc[(df['Care'] == 'CC_1')]
df_CC_2 = df.loc[(df['Care'] == 'CC_2')]
df_CC_3 = df.loc[(df['Care'] == 'CC_3')]



df_FC_1 = df.loc[(df['Care'] == 'FC_1')]
df_FC_2 = df.loc[(df['Care'] == 'FC_2')]
df_FC_3 = df.loc[(df['Care'] == 'FC_3')]
fig_dims = (12, 4)

fig, ax = plt.subplots(figsize=fig_dims)

sns.lineplot(x="Iteration", y="Score", data=df_FC_1, color="royalblue", markers=True, label="FC_1")
sns.lineplot(x="Iteration", y="Score", data=df_FC_2, color="lightskyblue", markers=True, label="FC_2")
sns.lineplot(x="Iteration", y="Score", data=df_FC_3, color="aqua", markers=True, label="FC_3")

sns.lineplot(x="Iteration", y="Score", data=df_CC_1, color="red", markers=True, label="CC_1")
sns.lineplot(x="Iteration", y="Score", data=df_CC_2, color="tomato", markers=True, label="CC_2")
sns.lineplot(x="Iteration", y="Score", data=df_CC_3, color="lightsalmon", markers=True, label="CC_3")
plt.legend()
plt.title("Head-Head FCvsCC Individual Outdegree Raw Scores")
# print(df)
```
